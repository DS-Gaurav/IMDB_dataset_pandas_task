```python
# 20Calculate average ratings per year
average_ratings = x.groupby('Year')['Rating'].mean()

# Create a line plot
plt.figure(figsize=(10, 6))
plt.plot(average_ratings.index, average_ratings.values, marker='o')
plt.title('Average Movie Ratings Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.grid(True)
plt.show()
```

```python
#21Identify trends in average ratings over the years
average_ratings = x.groupby('Year')['Rating'].mean()

# Create a line plot
plt.figure(figsize=(10, 6))
plt.plot(average_ratings.index, average_ratings.values, marker='o')
plt.title('Average Movie Ratings Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.grid(True)
plt.show()
```

```python
x['Actors'] = x['Actors'].str.split(', ')

# Explode the DataFrame to create a separate row for each actor
x = x.explode('Actors')

# Count the occurrences of each director-actor pair
director_actor_counts = x.groupby(['Director', 'Actors']).size().reset_index(name='Count')

# Find the most common director-actor pairs
most_common_pairs = director_actor_counts.sort_values(by='Count', ascending=False).head(5)

# Display the most common director-actor pairs
print("Most Common Director-Actor Pairs:")
print(most_common_pairs)
```

```python
# Calculate average runtime per year
average_runtime_per_year = x.groupby('Year')['Runtime (Minutes)'].mean()

# Create a line plot
plt.figure(figsize=(10, 6))
plt.plot(average_runtime_per_year.index, average_runtime_per_year.values, marker='o')
plt.title('Average Movie Runtimes Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Runtime (minutes)')
plt.grid(True)
plt.show()

# Calculate overall trend
overall_trend = 'increasing' if average_runtime_per_year.diff().mean() > 0 else 'decreasing'
print(f"Overall Trend: Movies are {overall_trend} in runtime")
```

```python
genre_counts_per_year = x.groupby(['Year', 'Genre']).size().unstack(fill_value=0)

# Plot the trends of different genres over the years
plt.figure(figsize=(10, 6))
genre_counts_per_year.plot()
plt.title('Genre Trends Over the Years')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
#plt.legend(title='Genre')
plt.grid(True)
plt.show()
```

```python
#Set the movie titles as the index for better visualization
x.set_index('Title', inplace=True)
# Create bar plots for each metric
plt.figure(figsize=(12, 6))

# Bar plot for ratings
plt.subplot(1,2,1)
x['Rating'].plot(kind='bar', color='blue')
plt.title('Movie Ratings')
plt.ylabel('Rating')

# Bar plot for revenue
plt.subplot(1,2,2)
x['Revenue (Millions)'].plot(kind='bar', color='green')
plt.title('Movie Revenue')
plt.ylabel('Revenue (Millions)')
plt.tight_layout()
plt.show()
```

```python
#28Analyze the relationship between the number of votes and movie ratings
plt.figure(figsize=(10, 6))
plt.scatter(x['Votes'], x['Rating'], color='blue', alpha=0.7)
plt.title('Number of Votes vs. Movie Ratings')
plt.xlabel('Number of Votes')
plt.ylabel('Rating')
plt.grid(True)
plt.show()

# Calculate the correlation coefficient
correlation = x['Votes'].corr(x['Rating'])
print("Correlation Coefficient:", correlation)
```

```python
# 29,Explore the distribution of ratings for different genres. Are there variations?
# Get unique genres
unique_genres = x['Genre'].unique()
# Create box plots to explore rating distribution for different genres
plt.figure(figsize=(10, 6))
plt.boxplot([x[x['Genre'] == genre]['Rating'] for genre in unique_genres], labels=unique_genres)
plt.title('Distribution of Ratings by Genre')
plt.xlabel('Genre')
plt.ylabel('Rating')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.show()
```

```python
#30 find outliers
# Calculate z-scores for movie runtimes
z_scores = (x['Runtime (Minutes)'] - x['Runtime (Minutes)'].mean()) / x['Runtime (Minutes)'].std()

# Identify outliers using z-scores
outliers = x[abs(z_scores) > 3]  # Adjust the threshold as needed

# Create a box plot to visualize runtime data
plt.figure(figsize=(6, 4))
plt.boxplot(x['Runtime (Minutes)'])
plt.title('Movie Runtime Box Plot')
plt.ylabel('Runtime (minutes)')
plt.show()

# Print the identified outliers
print("Identified Outliers:")
print(outliers)



```
